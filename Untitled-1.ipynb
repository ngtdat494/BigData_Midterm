```python
import pandas as pd
from pymongo import MongoClient
from urllib.parse import quote_plus

# ===== 1. Đọc dữ liệu từ Google Drive =====
file_id = "1fF0pfEZq05PAxhcPhjRjOUO5uP7gLRwn"
url = f"https://drive.google.com/uc?export=download&id={file_id}"

df = pd.read_csv(url)
print("Số dòng dữ liệu:", len(df))

records = df.to_dict(orient="records")

# ===== 2. Kết nối MongoDB Atlas =====
username = quote_plus("User")
password = quote_plus("123456@Zz")

uri = f"mongodb+srv://{username}:{password}@cluster0.8ugfq6t.mongodb.net/?retryWrites=true&w=majority"

client = MongoClient(
    uri,
    serverSelectionTimeoutMS=30000,
    connectTimeoutMS=30000,
    socketTimeoutMS=30000
)

# ===== 3. Chọn DB & Collection =====
db = client["covid_db"]
col = db["owid_covid_data"]

# ===== 4. Insert theo batch (ổn cho Colab) =====
batch_size = 500

for i in range(0, len(records), batch_size):
    col.insert_many(records[i:i+batch_size], ordered=False)
    print(f"Inserted {i + batch_size} records")

print("✅ Insert hoàn tất:", col.count_documents({}))
```
